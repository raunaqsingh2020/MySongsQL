```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read df from csv
raw_df = pd.read_csv("raw_tracks_features.csv")

raw_df
```

```python
# Find rows with invalid values
raw_df[raw_df.isna().any(axis=1)]
```

```python
# Get albums df
album_df = raw_df[["album_id", "album"]].drop_duplicates()

# Rename columns
album_df = album_df.rename(columns={"album_id": "id", "album": "name"})

# Check primary key constraint
album_df = album_df.set_index("id", verify_integrity=True)

print(album_df)
album_df.to_csv("albums.csv")
```

```python
# Get artists and song_artists dfs

from ast import literal_eval

# Gets unique (song_id, artist, artist_id) pairs
temp_df = raw_df[["id", "artists", "artist_ids"]].drop_duplicates()
    
# Using lists instead of df concat for O(N) instead of O(N^2) runtime
song_id_list = []
artist_id_list = []
artist_name_list = []
    
# Iteratively add all (song_id, artist, artist_id) pairs
for _, row in temp_df.iterrows():
    song_id = row["id"]
    artist_ids = literal_eval(row["artist_ids"])
    artist_names = literal_eval(row["artists"])
        
    if len(artist_ids) != len(artist_names):
        raise Exception(row["artist_ids"] + " and " + row["artists"] + " not the same length")
        
    for i in range(len(artist_ids)):
        song_id_list.append(song_id)
        
    artist_id_list.extend(artist_ids)
    artist_name_list.extend(artist_names)
            
# Create artists df
artist_df = pd.DataFrame.from_dict({"id": artist_id_list, "name": artist_name_list})

# Drop duplicate artist names (not case sensitive, but retain original case)
artist_df["name_upper"] = artist_df["name"].astype(str).str.upper()
artist_df = artist_df.drop_duplicates(subset=["id", "name_upper"], keep="first").drop("name_upper", axis=1)

# Check primary key constraint
artist_df = artist_df.set_index("id", verify_integrity=True)

# Save artists df to CSV
print(artist_df)
artist_df.to_csv("artists.csv")


# Create song_artists df
song_artist_df = pd.DataFrame.from_dict({"song_id": song_id_list, "artist_id": artist_id_list}).drop_duplicates()

# Save song_artists df to CSV
print(song_artist_df)
song_artist_df.to_csv("song_artists.csv", index=False)
```

```python
import re

# get songs df
song_df = raw_df.drop(columns=["album", "artists", "artist_ids"])

# Rename columns
song_df = song_df.rename(columns={"song_id": "id", "song": "name", "duration_ms": "duration"})

# Change boolean from string to int
song_df["explicit"] = song_df["explicit"].astype(int)

# Remove invalid dates
for indx, row in song_df.iterrows():
    if row["release_date"].count("-") != 2:
        song_df.loc[indx, "release_date"] = None
    
# Check primary key constraint
song_df = song_df.set_index("id", verify_integrity=True)

# Save to CSV
print(song_df)
song_df.to_csv("songs.csv")
```
