```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read df from csv
charts_df = pd.read_csv("raw_charts.csv")
```

```python
# Find rows with invalid values
charts_df[charts_df.isna().any(axis=1)]
```

```python
# Replace all invalid streams with 0
charts_df["streams"] = charts_df["streams"].fillna(0)

# Delete rows with an invalid title or artist
charts_df = charts_df.dropna(subset=['title', 'artist'])

# Check if there are still any invalid rows
charts_df[charts_df.isna().any(axis=1)]
```

```python
# Get song_id from url
def get_id(url):
    return url.rpartition("/")[-1]

charts_df["song_id"] = charts_df["url"].apply(get_id)

# Drop artist and title column
charts_df = charts_df.drop(["artist", "title"], axis=1)

charts_df
```

```python
# Check for duplicate (song_id, date, region, chart) pairs
results = charts_df.groupby(["song_id", "date", "region", "chart"]).size()
results = results[results > 1]
results
```

```python
# Save to csv
charts_df.to_csv("charts.csv", index=False)
```
